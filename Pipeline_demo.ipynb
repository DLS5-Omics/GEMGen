### Generate small molecules using the GEMGen generator

```python
import sys
sys.path.insert(0, "gemgen")
from generator import GEMGenGenerator
from utils import read_lines
from pathlib import Path
import json
import torch
gen = GEMGenGenerator(
    tokenizer_path="./models/generator/checkpoint1",
    ckpt_path="./models/generator/checkpoint1",
    tensor_parallel_size=1,
    pipeline_parallel_size=1,
    gpu_memory_utilization=0.8,
)

inputs = read_lines("./data/generator_prompts.txt")

outputs = gen.chat_batch(
    inputs,
    max_new_tokens=512,
    temperature=1.0,
    sample_count=100,
    sample_batch_size=10
)

result = []
for inp, out_list in zip(inputs, outputs):
    result.append([inp, out_list])

output_file = "./results/generator_results.json"
Path(output_file).parent.mkdir(parents=True, exist_ok=True)
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(result, f, ensure_ascii=False, indent=2)
print(f"Results saved to {output_file}")

for i, out_list in enumerate(outputs):
    print(f"Input: {inputs[i]}")
    print(f"Outputs: {out_list}")

torch.cuda.empty_cache()
```

### Score the generated molecules using the GEMGen scorer

```python
import sys
sys.path.insert(0, "gemgen")

from scorer import GEMGenScorer
import torch
import pandas as pd
import os
from pathlib import Path

DICT_PATH = "./models/scorer"
MODEL_PATH = "./models/scorer/model.safetensors"
# Simulate argparse args needed by GEMGenScorer
class Args:
    def __init__(self):
        self.dict_path = DICT_PATH
        self.tokenizer_path = DICT_PATH
        self.batch_size = 8
        self.max_len = 8192
        self.device = "cuda" if torch.cuda.is_available() else "cpu"

args = Args()
# Initialize scorer
print("Loading GEMGen Scorer...")
scorer = GEMGenScorer(args)

# Load model weights
print(f"Loading weights from {MODEL_PATH}...")
if MODEL_PATH.endswith('.safetensors'):
    from safetensors.torch import load_file
    state_dict = load_file(MODEL_PATH)
else:
    state_dict = torch.load(MODEL_PATH, map_location='cpu')

# Handle checkpoint format
if 'model_state_dict' in state_dict:
    scorer.load_state_dict(state_dict['model_state_dict'])
else:
    scorer.load_state_dict(state_dict)

scorer.to(args.device)
scorer.eval()
print("✅ Scorer loaded successfully.")

# Load generated molecules from previous evaluation step
input_csv = "./results/evaluated_molecules.csv"
print(f"Loading molecules from {input_csv}...")
df = pd.read_csv(input_csv)

descriptions = df['description'].astype(str)
smiles = df['smiles'].astype(str)

# Score all molecules
print(f"Scoring {len(smiles)} molecules...")
scores = scorer.score_samples(descriptions, smiles)

# Add scores to dataframe
df['score'] = scores

# Save updated results
output_csv = "./results/scored_molecules.csv"
Path(output_csv).parent.mkdir(parents=True, exist_ok=True)
df.to_csv(output_csv, index=False)
print(f"✅ Final results with scores saved to {output_csv}")

# Print summary
valid_scores = [s for s in scores if s is not None and not (isinstance(s, float) and pd.isna(s))]
if valid_scores:
    print(f"\nScore statistics (n={len(valid_scores)}):")
    print(f"  Mean: {sum(valid_scores)/len(valid_scores):.4f}")
    print(f"  Min:  {min(valid_scores):.4f}")
    print(f"  Max:  {max(valid_scores):.4f}")

print(df.head()[['description', 'smiles', 'score', 'QED', 'SAS', 'Ro5']].to_string())
```
